## 2.

Create a cleaned DataFrame `ufo`:

* Drop fully-empty or irrelevant columns (e.g., unnamed columns).
* Parse `datetime` to `datetime64[ns]` (`errors='coerce'`).
* Coerce `duration (seconds)`, `latitude`, `longitude` to numeric.
* Lowercase/trim `city`, `state`, `country`, `shape`.
* Remove rows with impossible coordinates (lat ∉ [-90,90], lon ∉ [-180,180]).
* Drop exact duplicates based on a reasonable subset (document your choice).
* Write at least 2 tests for your code (focus on the most complicated parts), then answer: What did you test for? How do you know your code is correct?


````python
ufo = ufo_raw.copy()
ufo = ufo.drop(columns=['Unnamed: 11'])

def parse_types(ufo):
    ufo = ufo.copy()
    ufo["datetime"] = pd.to_datetime(ufo["datetime"], errors="coerce")
    for col in ["duration (seconds)", "latitude", "longitude"]:
        ufo[col] = pd.to_numeric(ufo[col], errors="coerce")

    return ufo

def standardize_text(ufo):
    ufo = ufo.copy()
    for col in ["city", "state", "country", "shape"]:
        ufo[col] = ufo[col].str.lower().str.strip()

    return ufo

def filter_coordinates(ufo):
    ufo = ufo.copy()
    ufo = ufo[ufo["latitude"].between(-90, 90, inclusive="both")
        & ufo["longitude"].between(-180, 180, inclusive="both")]
    return ufo.reset_index(drop=True)

def drop_duplicates(ufo):
    ufo = ufo.copy()
    ufo = ufo.drop_duplicates(subset=["datetime", "city", "state", "country", "shape", "duration (seconds)", "latitude", "longitude"])
    return ufo.reset_index(drop=True)

ufo = parse_types(ufo)
ufo = standardize_text(ufo)
ufo = filter_coordinates(ufo)
ufo = drop_duplicates(ufo)

# Tests
ufo.info()
assert pd.api.types.is_datetime64_any_dtype(ufo["datetime"])
assert pd.api.types.is_numeric_dtype(ufo["duration (seconds)"])
assert pd.api.types.is_numeric_dtype(ufo["latitude"])
assert pd.api.types.is_numeric_dtype(ufo["longitude"])

# coords always valid
assert ufo["latitude"].between(-90, 90, inclusive="both").all()
assert ufo["longitude"].between(-180, 180, inclusive="both").all()

#check if any duplicates remain
dup_cols = ["datetime", "city", "state", "country", "shape", "duration (seconds)", "latitude", "longitude"]
duplicates = ufo.duplicated(subset=dup_cols)
assert not duplicates.any()
````

I performed sanity checks on all variables that were modified during the cleaning process. Specifically, I verified that the datetime column was converted to the datetime64[ns] type, and that duration (seconds), latitude, and longitude were successfully coerced to float values. I also tested that all latitude and longitude values fell within their valid ranges. Finally, to verify that duplicate UFO sightings were removed, I used the same set of variables chosen for deduplication to identify any remaining duplicate rows and confirmed that none were present. Together, these tests confirm that each cleaning step behaved as intended and that the resulting dataset satisfies the expected invariants.

````python
duration_hrs_mins = ufo["duration (hours/min)"].dropna().str.lower().str.strip()

def extract_min_like(duration_hrs_mins):
  duration_mins = duration_hrs_mins[
      duration_hrs_mins.str.contains(r"^\d+(\.\d+)?$|\bminute|\bmin\b|\bmins\b|\bmin\b|\b\d+\s*m\b",regex=True)]
  duration_mins.drop_duplicates(inplace=True)
  duration_mins.reset_index(drop=True, inplace=True)
  return duration_mins

duration_mins = extract_min_like(duration_hrs_mins)
print(duration_mins)
# Tests
test_set = pd.Series(
    ["several minutes","5 minutes","5 min","5 mins","4 hours","5 min.","10 m","6-7 min","approx. 5 min", "2 m", "~1 m", "2"])

output = extract_min_like(test_set).to_list()

assert "several minutes" in output
assert "5 minutes" in output
assert "5 min" in output
assert "5 mins" in output
assert "10 m" in output
assert "6-7 min" in output
assert "approx. 5 min" in output
assert "2 m" in output
assert "~1 m" in output
assert "2" in output
````

## 4.

Create additional columns in a new feature engineering version of the dataset called `ufo_fe`:

- `year`, `month`, `hour` from `datetime`
- `duration_log10` = log10(duration_seconds + 1)
- `duration_bucket` = categorical bins for `duration (seconds)` (you may adjust edges)
- `us_only` = 1 if `country == 'us'` else 0

Then produce:
- Value counts of `shape` (top 10).
- A table of sightings by `year` (counts).
- A `state × shape` table (top 10 states by sample size).
- A matplotlib or seaborn visualization of the top 10 states by number of UFO sightings. Write about which visualization you selected, and why.
- Write at least 2 tests for your code (focus on the most complicated parts), then answer: What did you test for? How do you know your code is correct?




````python
import numpy as np
import matplotlib.pyplot as plt


def extract_dt_features(ufo):
  ufo_fe = ufo.copy()
  ufo_fe["year"] = ufo_fe["datetime"].dt.year
  ufo_fe["month"] = ufo_fe["datetime"].dt.month
  ufo_fe["hour"] = ufo_fe["datetime"].dt.hour
  return ufo_fe

def calc_duration_log10(ufo):
  ufo_fe = ufo.copy()
  ufo_fe["duration_log10"] = np.log10(ufo_fe["duration (seconds)"] + 1)
  return ufo_fe

def add_duration_bucket(ufo):
  ufo_fe = ufo.copy()

  bins = [-1, 10, 60, 300, 600, 1800, 3600, 86400, np.inf]
  labels = ["<= 10 sec", "10–60 sec", "1-5 min", "5–10 min", "10-30 min", "30–60 min", "1-24 hr", "> 1 day"]

  ufo_fe["duration_bucket"] = pd.cut(ufo["duration (seconds)"], bins=bins, labels=labels)

  return ufo_fe

def find_us_only(ufo):
  ufo_fe = ufo.copy()
  ufo_fe["us_only"] = (ufo_fe["country"] == "us").astype(int)
  return ufo_fe

ufo_fe = extract_dt_features(ufo)
ufo_fe = calc_duration_log10(ufo_fe)
ufo_fe = add_duration_bucket(ufo_fe)

#Produce:
shape_counts = ufo_fe["shape"].value_counts().head(10)
print(shape_counts)
print()
sightings_by_year = ufo_fe.groupby("year").size().rename("count")
print(sightings_by_year)

top10_states = ufo_fe["state"].value_counts().head(10)
state_shape_table = (ufo_fe[ufo_fe["state"].isin(top10_states.index)].pivot_table
 (
  index="state",
  columns="shape",
  values="datetime",
  aggfunc="count",
  fill_value=0
  )
)
print(state_shape_table)

top_states = ufo["state"].value_counts().head(10)
top10_states.plot(kind="bar")
plt.xlabel("state")
plt.ylabel("num. UFO sightings")
plt.show()

# Tests

df_dates = pd.DataFrame({
  "datetime": pd.to_datetime(["2026-01-19 13:45:00", "2003-06-19 00:00:00"])
})
output = extract_dt_features(df_dates)

assert output.loc[0, "year"] == 2026
assert output.loc[0, "month"] == 1
assert output.loc[0, "hour"] == 13
assert output.loc[1, "year"] == 2003
assert output.loc[1, "month"] == 6
assert output.loc[1, "hour"] == 0

df_bins = pd.DataFrame({"duration (seconds)": [0, 11, 60, 299, 300, 600, 3600, 90000]})
out = add_duration_bucket(df_bins)

assert str(out.loc[0, "duration_bucket"]) == "<= 10 sec"
assert str(out.loc[1, "duration_bucket"]) == "10–60 sec"
assert str(out.loc[2, "duration_bucket"]) == "10–60 sec"
assert str(out.loc[3, "duration_bucket"]) == "1-5 min"
assert str(out.loc[4, "duration_bucket"]) == "1-5 min"
assert str(out.loc[5, "duration_bucket"]) == "5–10 min"
assert str(out.loc[6, "duration_bucket"]) == "30–60 min"
assert str(out.loc[7, "duration_bucket"]) == "> 1 day"
````

I decided to focus one task on geospatial clustering. Because it's hard to fully rely on the city, country and state responses because they are not entirely accurate, we should attempt to match and extract geospatial data from the latitude and longitude data as well. By creating clusters, this can also reveal information about the density of the locations with reported sightings, and we can see how they closely align with highly populated areas. I want to demonstrate the value of the clusters, so I also included a TX only dataset that is clustered so that we can more clearly see the clusters and how they align with what we know as cities, and how occasionally, there are clusters in places we don't expect.

````python
from sklearn.cluster import DBSCAN

coords = ufo[["latitude", "longitude"]].values
db = DBSCAN(eps=0.01, min_samples=50, metric="haversine").fit(coords)
ufo["geo_cluster"] = db.labels_

scatter = plt.scatter(
    ufo["longitude"],
    ufo["latitude"],
    c=ufo["geo_cluster"],
    s=5,
    alpha=0.5
)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geospatial Clusters of UFO Sightings")

plt.colorbar(scatter, label="Cluster Size")
plt.show()
````
